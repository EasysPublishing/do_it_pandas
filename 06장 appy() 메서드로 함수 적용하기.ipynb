## 06-3 람다 함수 사용하기 

#### [Do It! 실습] 데이터프레임에 람다 함수 사용하기

```python
df = pd.DataFrame({'a': [10, 20, 30],
                   'b': [20, 30, 40]})
print(df)
```

```python
def my_sq(x):
    return x ** 2

df['a_sq'] = df['a'].apply(my_sq)
print(df)
```

```python
df['a_sq_lamb'] = df['a'].apply(lambda x: x ** 2)
print(df)
```

## 06-4 벡터화된 함수 사용하기

#### [Do It! 실습] 벡터화된 함수 사용하기

```python
df = pd.DataFrame({"a": [10, 20, 30], 
                   "b": [20, 30, 40]})
print(df)
```

```python
def avg_2(x, y):
    return (x + y) / 2
```

```python
print(avg_2(df['a'], df['b']))
```

```python
import numpy as np

def avg_2_mod(x, y):
    if (x == 20):
        return(np.NaN)
    else:
        return (x + y) / 2
```

```python
# print(avg_2_mod(df['a'], df['b']))  # 오류
```

```python
print(avg_2_mod(10, 20))
```

```python
print(avg_2_mod(20, 30))
```

### 넘파이와 넘바로 벡터화하기

#### [Do It! 실습] 넘파이로 벡터화하기

```python
import numpy as np

avg_2_mod_vec = np.vectorize(avg_2_mod)
```

```python
print(avg_2_mod_vec(df['a'], df['b']))
```

```python
@np.vectorize
def v_avg_2_mod(x, y):
    if (x == 20):
        return(np.NaN)
    else:
        return (x + y) / 2

print(v_avg_2_mod(df['a'], df['b']))
```

#### [Do It! 실습] 넘바로 벡터화하기

```python
import numba

@numba.vectorize
def v_avg_2_numba(x, y):
    if (int(x) == 20):
        return(np.NaN)
    else:
        return (x + y) / 2
```

```python
# print(v_avg_2_numba(df['a'], df['b']))  # 오류
```

```python
print(v_avg_2_numba(df['a'].values, df['b'].values))
```
